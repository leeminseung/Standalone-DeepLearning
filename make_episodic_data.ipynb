```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```
pieces = []
compact_ratio = 1
columns = ['Current', 'Env1', 'Env2', 'Env3', 'Env4', 'Env5', 'Env6', 'Env7',  'Target', 'Agent']

for i in range(9):
    data_dir = '/content/drive/MyDrive/SKI_PBL/main_project/piece_{}.csv'.format(i) # piece data directory
    pieces.append(pd.read_csv(data_dir, index_col=0).loc[:, columns])
```

```
EPISODE_LEN = 240
COLUMN_NUM = 10
```

```
# np.array_split(pieces[1], len(pieces[1])//1440 + 1)
data = np.empty((0, EPISODE_LEN, COLUMN_NUM))
for i in range(len(pieces)):
    # print(len(pieces[i]))
    remain_rown = len(pieces[i]) - len(pieces[i])//EPISODE_LEN * EPISODE_LEN
    cat_array = np.array_split(pieces[i].iloc[:-remain_rown], len(pieces[i])//EPISODE_LEN)
    data = np.concatenate((data, cat_array), axis=0)

data.shape
```

```
np.save('/content/drive/MyDrive/SKI_PBL/main_project/episode_data', data)
```

## Data 결함
: target의 std가 대부분 0이므로 이부분을 고려해야함. (z normalization 불가)

```
cnt = 0
for i in range(data.shape[0]):
    # print(data[i,:,:].std(axis=0))
    if not 0. in data[i,:,:].std(axis=0):
        cnt += 1
    # print(data[i,:,:].mean(axis=0), data[i,:,:].std(axis=0))
print('not zero std of target %d' % cnt)
```
